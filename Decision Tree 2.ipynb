```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('shop data.csv')
df.head()
```

```python
x = df.iloc[:,:-1]
x
```

```python
y = df.iloc[:,4]
y
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_x = LabelEncoder()
x = x.apply(le_x.fit_transform)
```

```python
x
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model = model.fit(x,y)
```

```python
xinput = np.array([1,1,0,0])
```

```python
ypredict = model.predict([xinput])
ypredict
```
